```python
import pandas as pd
import numpy as np
import datetime
import os
```

```python
fs=[f for f in os.listdir('data') if '_data.csv' in f]
```

```python
def get_df(fn,start,end):
    df = pd.read_csv('data/{}'.format(fn))
    df.index = pd.to_datetime(df['time'])
    df=df.sort_index()
    g = df.groupby(df.index.date)
    to_return = []
    for k,v in g.groups.items():
        if k.month==7 and k.day>=10:
            #print(k)
            day_dict = df.loc[v]
            day_dict = day_dict.loc[(day_dict.index.hour>=start) & (day_dict.index.hour<=end)]
            to_return.append(day_dict)
    
    return to_return
```

```python
df = get_df(fs[0],7,12)
```

```python
def get_sedentary_fraction(df):
    
    return df.loc[(df['available']==1 )& (df['online_state']==1)].shape[0]/len(df.index)
```

```python
def get_lengths_counts(df):
    
    return len(df.index)
```

```python
def get_sedentary_count(df):
    
    return df.loc[(df['available']==1 )& (df['online_state']==1)].shape[0]
```

```python
def get_message_count(df):
    return df.loc[(df['available']==1 )& (df['action']==1)].shape[0]
```

```python
def get_message_fraction(df):
    denom = df.loc[(df['available']==1 )& (df['online_state']==1)].shape[0]
    if denom>0:
        return df.loc[(df['available']==1 )& (df['action']==1)& (df['online_state']==1)].shape[0]/ denom
    else:
        return 0
```

```python
def get_probaction_fraction(df):
    return [i for i in  df.loc[(df['available']==1 )&  (df['online_state']==1)]['probaction']]
```

```python
def get_all_user(fn,start,end):
    df = get_df(fn,start,end)
    
    fractions = [get_sedentary_fraction(d) for d in df if len(d.index)>0]
    sed_count = [get_sedentary_count(d) for d in df if len(d.index)>0]
    action_count = [get_message_count(d) for d in df if len(d.index)>0]
    lens = [get_lengths_counts(d) for d in df if len(d.index)>0]
    mf = [get_message_fraction(d) for d in df if len(d.index)>0]
    pa = [get_probaction_fraction(d) for d in df if len(d.index)>0]
    return fractions,sed_count,action_count,lens,mf,[item for sublist in pa for item in sublist]
    
```

```python
f,s,a = get_all_user(fs[0])
```

```python
def get_for_all_users(fs,start,end):
    fractions = []
    sed_counts =[]
    action_counts = []
    lens =[]
    num_users = []
    messages = []
    probactions = []
    for f in fs:
        
        ff,s,a,l,m,p = get_all_user(f,start,end)

        if len(ff)>0 and 'deryn_antised_data.csv' not in f:
            
            print(f)
            #print(ff)
            print(np.array(ff).mean())
            print(np.array(l).mean())
            print(np.array(s).mean())
            print(np.array(a).mean())
            print(np.array(m).mean())
            num_users.append(f)
            fractions.extend(ff)
            sed_counts.extend(s)
            action_counts.extend(a)
            messages.extend(m)
            lens.extend(l)
            probactions.extend(p)
    print(num_users)
    #print(len(num_users))
    return np.array(fractions),np.array(sed_counts),np.array(action_counts),\
np.array(lens),np.array(messages),np.array(probactions)
```
